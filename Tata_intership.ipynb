```python
import pandas as pd
import numpy as np
```

```python
delinquency = pd.read_csv("/content/delinquency_dataset_colab_tata.csv")
```

```python
delinquency.head()
```

```python
delinquency.isnull().sum()
```

```python
delinquency['Income']=delinquency['Income'].fillna(delinquency['Income'].median())
delinquency['Credit_Score']=delinquency['Credit_Score'].fillna(delinquency['Credit_Score'].median())
```

```python
delinquency['Loan_Balance']=delinquency['Loan_Balance'].fillna(delinquency["Loan_Balance"].median())
```

```python
delinquency.dtypes
```

```python
x = delinquency.drop(['Customer_ID','Employment_Status','Credit_Card_Type','Delinquent_Account','Location','Month_1','Month_2','Month_3','Month_4','Month_5','Month_6'],axis=1)
y=delinquency['Delinquent_Account']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_scale = StandardScaler()
```

```python
X_scale
```

```python
X_scaled=X_scale.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_trian,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=45,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

Logictic Regression

```python
model=LogisticRegression(class_weight='balanced')
```

```python
model
```

```python
model.fit(x_trian,y_train)
```

```python
y_pred_log = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,f1_score,roc_auc_score,accuracy_score
```

```python
confusion_matrix(y_test,y_pred_log)
```

```python
f1_log=f1_score(y_test,y_pred_log)
```

```python
auc_log=roc_auc_score(y_test,y_pred_log)
```

```python
acc_log=accuracy_score(y_test,y_pred_log)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf=RandomForestClassifier(class_weight='balanced')
```

```python
model_rf
```

```python
model_rf.fit(x_trian,y_train)
```

```python
# Step 1: Get probabilities
y_prob_rf = model.predict_proba(x_test)[:,1]

# Step 2: Apply custom threshold
y_pred_rf = (y_prob_rf > 0.5).astype(int)
```

```python
confusion_matrix(y_test,y_pred_rf)
```

```python
auc_rf = roc_auc_score(y_test,y_prob_rf)
```

```python
f1_rf=f1_score(y_test,y_pred_rf)
```

```python
acc_rf=accuracy_score(y_test,y_pred_rf)
```

```python
from xgboost import XGBClassifier
```

```python
model_xgb = XGBClassifier(n_estimators=200,max_depth=5,random_state=45)
```

```python
model_xgb
```

```python
model_xgb.fit(x_trian,y_train)
```

```python
y_prob_xgb = model_xgb.predict_proba(x_test)[:,1]

y_pred_xgb = (y_prob_xgb > 0.1).astype(int)   # start with 0.4 threshold
```

```python
auc_xgb=roc_auc_score(y_test,y_prob_xgb)
```

```python
f1_xgb=f1_score(y_test,y_pred_xgb)
```

```python
acc_xgb=accuracy_score(y_test,y_pred_xgb)
```

```python
confusion_matrix(y_test,y_pred_xgb)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_gbr = GradientBoostingClassifier(n_estimators=200,max_depth=10,random_state=45)
```

```python
model_gbr.fit(x_trian,y_train)
```

```python
y_prob_gbc = model_xgb.predict_proba(x_test)[:,1]

y_pred_gbc = (y_prob_gbc > 0.1).astype(int)   # start with 0.4 threshold
```

```python
acc_gbc=accuracy_score(y_test,y_pred_gbc)
```

```python
confusion_matrix(y_test,y_pred_gbc)
```

```python
f1_gbc=f1_score(y_test,y_pred_gbc)
```

```python
auc_gbc=roc_auc_score(y_test,y_prob_gbc)
```

```python
metrics_df = pd.DataFrame({
    "Model": ["Logistic", "Random Forest", "XGBoost", "Gradient Boost"],
    "Accuracy": [acc_log, acc_rf, acc_xgb, acc_gbc],
    "F1 Score": [f1_log, f1_rf, f1_xgb, f1_gbc],
    "AUC": [auc_log, auc_rf, auc_xgb, auc_gbc]
})
```

```python
metrics_df
```

```python
metrics_df.to_csv("model_metrics.csv", index=False)
```
